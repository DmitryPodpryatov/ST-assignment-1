```python
Ns = [100_000 * 10 ** n for n in range(5)]
memory_profiling = [31, 35, 69, 412, 3845]  # MiB
time_profiling = [0.001, 0.002, 0.002, 0.002, 0.002]  # s
```

```python
import math


def actual_complexity(N):
    epsilon = 0.01
    log = math.log(epsilon * N)

    return 1 / epsilon * log * log


def nlogn(N):
    return N * math.log(N)
```

```python
import matplotlib.pyplot as plt

plt.xscale('log')
plt.yscale('linear')
plt.plot(Ns, memory_profiling, label='Benchmarks')
plt.plot(Ns, [actual_complexity(N) for N in Ns], label='Actual')
plt.legend()
plt.show()
```

```python
plt.xscale('log')
plt.yscale('linear')
plt.plot(Ns, memory_profiling, label='Benchmarks')
plt.plot(Ns, [1 / 10 * actual_complexity(N) for N in Ns], label='Actual with coefficient')
plt.legend()
plt.show()
```

```python
plt.xscale('log')
plt.yscale('linear')
plt.plot(Ns, memory_profiling, label='Benchmarks')
plt.plot(Ns, [1 / 10 * actual_complexity(N) for N in Ns], label='Actual with coefficient')
plt.plot(Ns, [nlogn(N) for N in Ns], label='N log N')
plt.legend()
plt.show()
```
